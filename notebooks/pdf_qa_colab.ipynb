```python
import re

def looks_like_real_text(text, min_words=10):
    words = re.findall(r'[a-zA-Z]{3,}', text)  # words with 3+ letters
    return len(words) >= min_words

all_text = " ".join([p.page_content for p in pages])
if not looks_like_real_text(all_text):
    print(" This looks like a scanned PDF with no real extractable text.")
else:
    print(" Real text detected.")
```
