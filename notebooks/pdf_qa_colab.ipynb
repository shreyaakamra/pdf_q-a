```python
filename = list(uploaded.keys())[0]

from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("shreya40416403223.pdf")
pages = loader.load()

total_chars = sum(len(p.page_content.strip()) for p in pages)
print(f"Total extracted characters: {total_chars}")

if total_chars < 50:
    print(" Warning: This looks like a scanned PDF with no extractable text.")
print(f"Total extracted characters: {total_chars}")
print("---ACTUAL EXTRACTED TEXT---")
print(pages[0].page_content)
```

```python
import re

def looks_like_real_text(text, min_words=10):
    words = re.findall(r'[a-zA-Z]{3,}', text)  # words with 3+ letters
    return len(words) >= min_words

all_text = " ".join([p.page_content for p in pages])
if not looks_like_real_text(all_text):
    print(" This looks like a scanned PDF with no real extractable text.")
else:
    print(" Real text detected.")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=80
)

chunks = splitter.split_documents(pages)

print(f"Number of chunks created: {len(chunks)}")
print("---First chunk preview---")
print(chunks[0].page_content)
print("---second chunk preview---")
print(chunks[1].page_content)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Quick test — embed a single sentence to confirm it works
test_vector = embeddings.embed_query("This is a test sentence")
print(f"Vector length: {len(test_vector)}")
print(test_vector[:5])
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(chunks, embeddings)

print("Vector store created successfully")
print(f"Number of vectors stored: {vectorstore.index.ntotal}")
```

```python
query = "What is the Windows API?"
results = vectorstore.similarity_search(query, k=3)

for i, doc in enumerate(results):
    print(f"--- Result {i+1} ---")
    print(doc.page_content)
    print()
```

```python
test_queries = [
    "What is the Windows API?",
    "What is the difference between user mode and kernel mode?",
    "What are the common features of an IDE?",
    "What is the consistency principle in UI design?",
    "What is error prevention?"
]

for q in test_queries:
    print(f"=== Query: {q} ===")
    results = vectorstore.similarity_search(q, k=2)
    for doc in results:
        print(doc.page_content[:250])
        print("---")
    print()
```
